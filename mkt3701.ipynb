```python
data = data.fillna(data['Churn'].value_counts().index[0])
data = data.fillna(data['StreamingTV'].value_counts().index[0])
data = data.fillna(data['StreamingMovies'].value_counts().index[0])
data = data.fillna(data['Contract'].value_counts().index[0])
print(data.isnull().values.sum())
```

```python
data_cleanup = {"Churn":     {"Yes": 1, "No": 0},
                "StreamingTV": {"Yes": 1, "No": 0, "No internet service": 0},
                "StreamingMovies": {"Yes": 1, "No": 0, "No internet service": 0}}
data.replace(data_cleanup, inplace=True)
```
