```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import numpy.random as rd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from scipy import stats

data = pd.read_csv("HomworkData.csv")                # Read the dataset
data.head(10)
```

```python
churn_binary = {"Churn": {"Yes":1, "No":0}}
data.replace(churn_binary, inplace=True)

techsupport_dummy = pd.get_dummies(data, columns=["TechSupport"])
pd.get_dummies(data, columns=["TechSupport"]).head()
```

```python
data = pd.concat([techsupport_dummy],axis=1)
```

```python
data.head()
```

```python
X = data[data.columns[-3::]]
Y = data["Churn"]

print(X)
print(Y)
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(Y,X)
result=logit_model.fit()
print(result.summary2())
```
